**Listas**

Es el equivalente a una matriz. Y para definir una lista tenemos que definir un conjunto de valores separados por comas y empezar por **[** y terminar por **]**

1. Puede contener cualquier valor tanto primitivos como objetos
2. Se puede acceder a un elemento a traves de su indice basado en base 0
3. Pueden tener varias dimensiones
4. Se puede insertar,borrar, modificar cualquier elemento de la lista
5. Una lista es del tipo **list**

**Una lista puede contener cualquier valor**

```
a=[1,1.1,"Hello",False]
a
```

**Acceder a un elemento de una lista**

```
a=[1,2,3]
print(a[0])
print(a[1])
print(a[2])
print(a[3])
```

**Lista de dos dimensiones**

```
a=[[1,2],[3,4],[5,6]]
print(a[0][1]) #devuelve el 2 en la primera dimension
print(a[2][0]) #devuelve el 5 en la tercera dimension
```

**Operaciones con listas**

```
a=[1,2,3]
a.append(4) #inserta el 4 en la última posición
a.insert(0,0) #inserta en la posición 0 el 0
a.remove(4) #elimina la primera ocurrencia del valor 4
a.pop() #elimina el ultimo valor de la lista
```

**Contar el numero de ocurrencias de un valor en una lista**

```
a=[1,1,2]
a.count(1)
```

**Ordenar una lista de forma ascendente**

```
a=[3,2,1]
a.sort()
a
```

**Ordenar una lista de forma descendente**

```
a=[1,2,3]
a.sort(reverse=True)
a
```

**Sumar dos listas con el operador +**

```
a=[1,2,3]
b=[4,5,6]
a+b
```

**Clonar los valores de una lista n veces**

```
a=[1,2,3]
a*=3
a
```

# Ejercicios

Imprimir  los pares de la siguiente lista [1,2,3]

```
def evens(L):
  for v in L:
    if(v % 2 == 0):
      print(v)
evens([1,2,3])
```

Escribir una función filter que nos permita filtrar por cualquier criterio cualquier lista

```
def filter(L,predicate):
  for v in L:
    if predicate(v):
      print(v)
filter([1,2,3,4,5],lambda v:v%2==0 or v>3) #imprimir pares y los mayores de 3
```
