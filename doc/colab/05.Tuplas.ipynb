**Tuplas**

Es el equivalente a una matriz inmutable. Y para definir una tupla tenemos que definir un conjunto de valores separados por comas y empezar por **(** y terminar por **)**

1. Puede contener cualquier valor tanto primitivos como objetos
2. Se puede acceder a un elemento a traves de su indice basado en base 0
3. Pueden tener varias dimensiones
4. No se puede insertar,borrar y modificar cualquier elemento de la lista, por ser imutable
5. Una Tupla es del tipo **tuple**

**Una tupla puede contener cualquier valor**

```
a=(1,1.1,"Hello",False)
b=1,1.1,"Hello",False
print(a)
print(b)
```

**Acceder a un elemento de una tupla**

```
a=(1,2,3)
print(a[0])
print(a[1])
print(a[2])
print(a[3])
```

**Tupla de dos dimensiones**

```
a=((1,2),(3,4),(5,6))
print(a[0][1]) #devuelve el 2 en la primera dimension
print(a[2][0]) #devuelve el 5 en la tercera dimension
```

**Operaciones con Tuplas**

```
a=(1,2,3)
a.append(4) #inserta el 4 en la última posición
a.insert(0,0) #inserta en la posición 0 el 0
a.remove(4) #elimina la primera ocurrencia del valor 4
a.pop() #elimina el ultimo valor de la lista
```

**Contar el numero de ocurrencias de un valor en una tupla**

```
a=(1,1,2)
a.count(1)
```

**Una tupla no se puede ordenar**

```
a=list((3,2,1))
a.sort()
a
```

**Sumar dos tuplas con el operador +**

```
a=(1,2,3)
b=(4,5,6)
a+b
```

**Clonar los valores de una tupla n veces**

```
a=(1,2,3)
a*=3
a
```

# Ejercicios

Iterar por una tupla que representa una linea

```
def line(T):
  for point in T:
    x,y = point
    print(x)
    print(y)


line(((0,2),(100,2)))
```
