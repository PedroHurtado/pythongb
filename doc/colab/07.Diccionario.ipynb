**Diccionario**

Es una estructura de datos **key**:**value** donde la clave no puede estar duplicada, equivalente a un map en otros lenguajes. Se define empezando por **{** y terminando por **}**, cada par key:value va separado por comas 

1. Puede contener cualquier valor tanto primitivos como objetos
2. Se puede acceder a un elemento a traves de su indice basado en su key
3. Un diccionario puede tener otros diccionarios anidados
4. Se puede insertar,borrar, modificar cualquier elemento del diccionario 
5. Un diccionario es del tipo **dict**

**Un diccionario puede contener cualquier valor**

```
teacher={'name':"Pedro", 'email': "pedrohurt@gmail.com","phones":{
    "home":616647015,
    "office":616647015
}}
teacher
```

**Acceder a un valor de un diccionario por su clave**

```
teacher={'name':"Pedro", 'email': "pedrohurt@gmail.com","phones":{
    "home":616647015,
    "office":616647015
}}
print(teacher['name'])
print(teacher['phones']['home'])
```

**Crear un diccionario y agregarle un nuevo valor**

```
a={}
a.update(id=1)
print(a)
```

**Modificar el valor de un diccionario**

```
a={}
a.update(id=1)
print(a)
a['id']=10
print(a)
```

**Eliminar una clave de un diccionario**

```
a={}
a.update(id=1)
print(a)
del a['id']
print(a)
```

**Comibnar dos diccionarios**

```
a={'id':1,'name':'Pedro'}
b={'name':'Pedro Hurtado'}
b.update(a)
b
```

Iterar por un diccionario y obtener la clave y su valore

```
def iter_dictionary(D):
  for k,v in D.items():
      print(k,"->",v)
iter_dictionary({'id':1})
```
