**Los tipos primitivos de python son los siguientes**


*   Enteros(int)
*   Coma flotante(float)
*   Cadenas(str)
*   Booleanos(bool)




**Enteros**

Representa cualquier numero negativo o positivo, pero sin decimales  

  * -1
  * 0
  * 1

```
print(type(-1))
print(type(0))
print(type(1))
```

**Coma Flotante**

Representa cualquier numero negativo o positivo con decimales como
  * -1.0
  * 0.0
  * 1.0

```
print(type(-1.0))
print(type(0.0))
print(type(1.0))
```

**Cadenas**

Representa cualquier conjunto de caracteres como "Hello" o bien una cadena vacia y su valor debe de estar delimitados por comillas dobles(") o bien comilla simple(')

Use comillas dobles para representar un cadena de lenguaje natural y simples para un literal(recomendación), pero sigue siempre la misma regla en todo tu código, aunque tal y como dice la [guia](https://www.python.org/dev/peps/pep-0008/#string-quotes) ambas son iguales.



  * "Hello"
  * 'Hello'
  * ""
  * ''

Cualquier secuencia de numeros delimitado por comillas dobles(") o simples(') tambien es una cadena.

```
print(type("Hello"))
print(type('Hello'))
print(type(""))
print(type(''))
print(type("123"))
print(type("123.00"))
```

**Booleanos**

Representan los valores verdadero y Falso

* True
* False

```
print(type(True))
print(type(False))
```

# Conversión entre tipos primitivos

**Conversion a booleanos**

Cualquiera de los tipos primitivos str,int y float se pueden convertir a booleano

```
print(bool("False")) #resultado True
print(bool("Hello")) #resultado True
print(bool("")) #resultado False
```

```
print(bool(0)) #False
print(bool(1)) #True
print(bool(-1)) #True
```

```
print(bool(0.1)) #False
print(bool(1.0)) #True
print(bool(-1.0)) #True
```

**Conversión a entero**



1.   Se puede convertir de cadena a entero siempre que una cadena solamente contenga valores numéricos.
2.   Se pueden convertir cualquiera de los valores booleanos(True,False) a enteros. True se convierte a 1 y False se convierte a 0
3.   Se puede convertir cualquier valor de coma flotante a entero con perdida de datos. 1.0 no respresenta perdida de datos y 1.1 o 1.6 devuelven 1 con la consiguiente perdida de datos de la parte decimal. Un error es pensar que cualquier decimal mayor de 0.5 devuelve 1, para ello tenemos de utilizar la funcion integrada [round](https://www.w3schools.com/python/ref_func_round.asp#:~:text=The%20round()%20function%20returns,will%20return%20the%20nearest%20integer.)



```
#conversion de cadena a entero
print(int("1234"))
print(int("Hello"))
```

```
#conversion de booleano a entero
print(int(True))
print(int(False))
```

```
#conversion de coma flotante a entero
print(int(1.0))
print(int(1.1))
print(int(1.6))
print(int(round(1.6)))
```

**conversion a coma flotante**

1.   Se puede convertir de cadena a coma flotante siempre que una cadena solamente contenga valores numéricos o bien numeros separando la parte entera de la parte demicimal por el caracter "."
2.   Se pueden convertir cualquiera de los valores booleanos(True,False) a coma flotante. True se convierte a 1.0 y False se convierte a 0.0
3.   Cualquier numero entero se puede convertir a coma flotante, siendo siempre la parte decimal 0. Si convertimos 0 como entero a coma flotante devuelve 0.0

```
#conversion de cadena a coma flotante
print(float("1234"))
print(float("1234.25"))
print(float("Hello"))
```

```
#conversion de booleano a coma flotante
print(float(True))
print(float(False))
```

```
print(float(-1))
print(float(0))
print(float(1))
```

**conversión a cadena**

Cualquiera de los tipos primitivos booleano, entero o como flotante se pueden convertir a cadena.

1. En el caso de los números enteros una secuencia de caracteres númericos.
2. En el caso de los números con coma flotante la parte entera y la parte decimal serán secuencias de numeros separadas por "."
3. En el caso de un tipo Booleano devolvera las cadenas "True" o "False"


```
#secuencia de caracteres numericos. conversion de entero a cadena
for v in str(1234):
  print(v)
```

```
#secuencia de caracteres numericos y el punto decimal. conversion de float a cadena
print("El importe de la factura es " + str(1234.25))

for v in str(1234.25):
  print(v)
```

```
#conversion booleano a cadena
print(str(True))
print(str(False))
```
