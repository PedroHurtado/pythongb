1. Operadores de asignacion

**Operadores de asignacion**

3. **-=** Resta el valor a una variable
4. ***=** Multiplica el valor de una variable
5. **/=** Divide el valor de la variable
6. ****=** Calcula el exponente de una variable
7. **//=** Calcula la división entera de una variable
8. **%=**  Calcula el resto de una variable

```
a=1
a+=10
a-=1
a*=10
a/=10
a**=6
a//=10
a%=10
a
```

**Operadores relacionales**

Todos estos operadores devuelven un tipo booleano que puede ser **True o False**

1. **==** Compara dos primitivos o dos variables y devuelve True si son iguales y False si son distintos
1. **!=** Compara dos primitivos o dos variables y devuelve True si son distintos o False si son iguales
3. **>** Compara dos primitivos o dos variables y devuelve True si a>b  o False si b<a
4. **>=** Compara dos primitivos o dos variables y devuelve True si a>=b  o False si b<=a
4. **<** Compara dos primitivos o dos variables y devuelve True si a<b  o False si b<a
4. **<=** Compara dos primitivos o dos variables y devuelve True si a<=b  o False si b<=a



```
a=1
b=2
print(a==b)
print(a>b)
print(b>a)
print(b>=a)
print(a<=b)
```

**Operadores lógicos**

Las igual que con los operadores relaciones el valor devuelto es True o False.

Se utilizan para combinar varios operadores relacionales

1. **and** Devuelve True si todos los operadores relacionales son verdaderos
2. **or**  Devuleve True si alguno de los operadores relacionales es verdadero
3. **not** Devuelve el valor contrario. Es decir si es verdadero devuelve False y si es falso devuelve True

```
#and
print(True and True)
print(True and False)
print(False and True)
print(False and False)
#or
print(True or True)
print(True or False)
rint(False or True)
print(False or False)p
#not
print(not True)
print(not False)
```

**Operador is**

Suele haber confusión con el operador == mientras que el operador == devuelve True si sus valores son iguales el operador is devuelve true si los dos objetos son iguales. Se puede utilizar con **not** para negarlo

```
a=[1,2,3]
b=[1,2,3]
a=b
print(a==b)
print(a is b)
c=b
print(c is b)
```

**Operador in**

Devuelve True si un valor esta presente en una secuencia, se puede utilizar en combinación con **not** para negarlo

```
print(1 in [1,2,3])
print(4 in [1,2,3])
print(4 not in [1,2,3])
```
