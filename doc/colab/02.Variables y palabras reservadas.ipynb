**Variable**

Una variable en cualquier lenguaje de programación es un identificador que es capaz de asignar o recuperar un valor guardado en una dirección de memoria.

**Restricciones en el nombre de las variables**
1.   Solo pueden contener caracteres ( A-Z, a-z), (0,9) y _
2.   El caracter incial de una variable solo puede empezar por el ( A-Z, a-z) o _
3.   En python es recomendable que las funciones esten en minúsculas y las clases empiecen por mayúscula aunque los tipos basicos como **int,bool,float,str** al igual que **object** de quien hereda todo(**en Python todo son objetos**) están escritas en minúsculas  [naming convention](https://www.python.org/dev/peps/pep-0008/#function-and-variable-names)
4.    Una constante, aunque python no las soporta , se debe de definir en Máyusculas (PI:3.1416). Estas variables deben de definirse a nivel de módulo.
5.    El nombre de las variables debe de definir su proposito. Evita variables como a,a1,a2,etc. [clean code](https://www.amazon.es/Clean-Code-Handbook-Software-Craftsmanship/dp/0132350882)
6.    Las variabes tienen un ambito global excepto en las funciones que se destruyen cuando acaba el bloque de código
7.    Al igual que una variable es creada con un nombre e inicializada con la sigiente **sintaxis a=1** tambien puede ser destruida
8.    Multiples variables pueden ser asignadas en una sola linea de código

**Ambito de variables**

```
a=10   #variable global no recomendado
def fn():
  a=23
  print(a)
fn()
print(a)
```

```
a=10 #variable gloabal
def fn():
  a=11  #variable privada a una funcion
  print(a)
fn()
print(a)
```

```
for v in [1,2,3]:
  print(v)
print(v)
```
