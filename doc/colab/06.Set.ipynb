**Set**

Secuencia de datos sin duplicados y desordenada.

Para crear una **Set** temenos que escribir una secuencia de valores separados por comans y delimitados por **{** y terminar por **}** o bien un set vacio de la siguiente forma set() o bien a partir de un iterable por ejemplo una lista set([1,2,3])

1. Puede contener cualquier valor tanto primitivos como objetos
2. No se puede acceder a un elemento a traves de su indice basado en base 0
3. No pueden tener varias dimensiones
4. Se pueden insertar,borrar y modificar
5. Un Set es del tipo **set**

**Un set puede contener cualquier valor**

```
a={1,"Hello", True}
a
```

**No se puede acceder a un elemento por su indice**

```
a={3,2,1}
a[2]
```

**No podemos tener set anidados**

```
a={1,2,{3}}
```

**Operaciones con Set**

```
a=set()
a.add(1)
a.add(2)
a.add(3)
a.add(4) #inserta el 4
a.pop() #elimina un elemento arbitrario y lo devuelve
a.update({4,5}) #modifica un set con otro set
a.remove(4) #elimina la primera ocurrencia del valor 4
a
```

**Los set no están ordenados para ello podemos convertir a una lista y después ordenar**

```
a={3,2,1}
b=list(a)
b.sort()
b
```

**Intetersección de dos set**

```
a={1,2,3}
b={3,5,6}
a.intersection(b)
```

**Interseccion con el operador &**

```
a={1,2}
b={3,5,6}
a&b
```

**Unión de dos set**

```
a={1,2,3}
b={3,4,5}
a.union(b)
```

**Unión de dos set con el operador |**

```
a={1,2,3}
b={3,4,5}
a|b
```

# Ejercicios

Imprimir  los pares del siguiente Set [1,2,3]

```
def evens(L):
  for v in L:
    if(v % 2 == 0):
      print(v)
evens({1,2,3})
```

Escribir una función filter que nos permita filtrar por cualquier criterio de un Set

```
def filter(L,predicate):
  for v in L:
    if predicate(v):
      print(v)
filter({1,2,3,4,5},lambda v:v%2==0 or v>3) #pares y los mayores de 3
```
